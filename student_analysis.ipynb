```python
#Visualizations
# istogram of Grades

Histograms help visualize the distribution of student grades across different assessment periods.
```

```python
import matplotlib.pyplot as plt

for grade in ['G1', 'G2', 'G3']:
    plt.figure(figsize=(6,4))
    plt.hist(studentmat_df[grade], bins=10)
    plt.title(f'Histogram of {grade}')
    plt.xlabel('Grade')
    plt.ylabel('Frequency')
    plt.show()
```

#Study Time vs Grades
Scatter plots help examine the relationship between the amount of study time and student grades.

```python
import matplotlib.pyplot as plt

# G1
plt.scatter(studentmat_df['studytime'], studentmat_df['G1'])
plt.title('Study Time vs G1')
plt.xlabel('Study Time')
plt.ylabel('G1')
plt.show()

# G2
plt.scatter(studentmat_df['studytime'], studentmat_df['G2'])
plt.title('Study Time vs G2')
plt.xlabel('Study Time')
plt.ylabel('G2')
plt.show()

# G3
plt.scatter(studentmat_df['studytime'], studentmat_df['G3'])
plt.title('Study Time vs G3')
plt.xlabel('Study Time')
plt.ylabel('G3')
plt.show()
```
